```python
#how rto drop null values
df.dropna(inplace=True)
```

```python
age_row=df[df['age'].isnull()]
age_row
```

```python
# ## not related to above but if we get replace option we are replacing the missing values by
# if the datatype  of columns object=replace missing values with ode of that column



# if the dtatatype of column is numeric=replace missing values with mean of that column
```

```python
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=df[i].fillna(df[i].mode()[0]) #hats why 0 is used  it selects the first mode value from the Series returned by .mode() to fill missing categorical values.
  else:
    df[i]=df[i].fillna(df[i].mean())
```

```python
#2) checking for duplicate values
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
#df.dropduplicates(inplace=True) right now no duplicates in data
```

```python
#3) check for outliers (optional     step)
import matplotlib.pyplot as plt
import seaborn as sns
for i in df.columns:
  if df[i].dtypes=='int64' or df[i].dtypes=='float64':
    plt.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```

```python
#how to remove the outliers:
for i in df.columns:
  if df[i].dtypes=='int64' or df[i].dtypes=='float64':
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    #put a condition that if any data that is outside the upper range and lower range
    #only consider the data between upper and lower range.
    df=df[(df[i]>=lower_limit) & (df[i]<=upper_limit)]
```

```python
#3) check for outliers (optional     step) now outliers are removed by above  sep
import matplotlib.pyplot as plt
import seaborn as sns
for i in df.columns:
  if df[i].dtypes=='int64' or df[i].dtypes=='float64':
    plt.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```
